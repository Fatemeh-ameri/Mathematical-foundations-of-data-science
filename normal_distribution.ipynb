normal distribution for continues variables(single real-valued)

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal
```

```python
SQRT_TWO_PI = math.sqrt(2 * math.pi)
```

```python
def normal_pdf(x: float, mu: float = 0, sigma: float = 1) -> float:
    return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (SQRT_TWO_PI * sigma))
```

```python
x = int(input("How many numbers do you want to enter?"))
data = []
for i in range(x):
    xs = float(input("enter your number"))
    data.append(xs)
mu = float(input("enter your mu")) 
sigma = float(input("enter your sigma"))   
```

```python
xs = [data / 10.0 for data in range(-50, 50)]
print([normal_pdf(data, sigma=sigma, mu = mu) for data in xs])
plt.plot(xs,[normal_pdf(data,sigma=sigma) for data in xs])
plt.legend()
plt.title("normal distribution")
plt.show()
```

normal distribution for continues variables(D-dimensional)

```python
random_seed=50
cov_val = [-0.8, 0, 0.8]
mean = np.array([0,0])
```

```python
for val in cov_val:
    cov = np.array([[1, val], [val, 1]])
    distr = multivariate_normal(cov = cov, mean = mean, seed = random_seed)
    data = distr.rvs(size = 50)
print(data)
```
